```python
### Решающий пень

import sys


def main():
    class MeanCalc:
        def __init__(self):
            self.count = 0
            self.mean = 0

        def Add(self, value, weight = 1.):
            self.count += weight
            self.mean += weight * (value - self.mean) / self.count

        def Remove(self, value, weight = 1.):
            self.Add(value, -weight)

    class MSECalculator:
        def __init__(self):
            self.MeanCalc = MeanCalc()
            self.MSE = 0.

        def Add(self, value, weight = 1.):
            curDiff = value - self.MeanCalc.mean
            self.MeanCalc.Add(value, weight)
            self.MSE += weight * curDiff * (value - self.MeanCalc.mean)

        def Remove(self, value, weight = 1.):
            self.Add(value, -weight)

    class Instances:
        def __init__(self):
            self.Items = []
            self.OverAllMSE = MSECalculator()

        def Read(self):
            n = int(input())

            for i in range(n):
                x, y = map(float, input().strip().split())
                self.Items.append([x, y])
                self.OverAllMSE.Add(y)
            self.Items.sort()
        
        def BestSplit(self):
            left = MSECalculator()
            right = self.OverAllMSE

            bestA = 0
            bestB = right.MeanCalc.mean
            bestC = self.Items[0][0]

            bestQ = right.MSE

            for i in range(len(self.Items) - 1):
                item = self.Items[i]
                next_item = self.Items[i + 1]

                left.Add(item[1])
                right.Remove(item[1])
    
                if item[0] == next_item[0]:
                    continue 

                a = left.MeanCalc.mean
                b = right.MeanCalc.mean
                c = (item[0] + next_item[0]) / 2

                q = left.MSE + right.MSE

                if q < bestQ:
                    bestA, bestB, bestC, bestQ = a, b, c, q
            return bestA, bestB, bestC

    instances = Instances()
    instances.Read()
    a, b, c = instances.BestSplit()

    print('{:.7f} {:.7f} {:.7f}'.format(a, b, c))
    

if __name__ == '__main__':
    main()
```
